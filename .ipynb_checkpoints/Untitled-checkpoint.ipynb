```python
import pandas as pd
import csv
import numpy as np
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

```python
def fichier_csv(file_aspx, file_csv):
    
  # Lire le fichier .aspx en utilisant read_html
  tables = pd.read_html(file_aspx)
  # Sélectionner la première table dans la liste
  df = tables[0]
  #
  columns=['date','closing','adjusted','evolution','quantity','volume']
  #  
  df.columns= columns
  #  
  df = df.drop(index=0) 
  # Enregistrer le DataFrame en tant que fichier .csv
  df.to_csv(file_csv, index=False) 
```

```python
fichier_csv('Data_BMCI.aspx', 'fichier_BMCI.csv')
fichier_csv('Data_Attijariwafa_bank.aspx', 'fichier_Attijariwafa_bank.csv')
fichier_csv('Data_Bank_of_Africa.aspx', 'fichier_Bank_of_Africa.csv')
fichier_csv('Data_BCP.aspx', 'fichier_BCP.csv')
fichier_csv('Data_CIH.aspx', 'fichier_CIH.csv')
```

```python
#date,closing,adjusted,evolution,quantity,volume
data = pd.read_csv('fichier_Bank_of_Africa.csv')
```

```python
data.tail(10)
```

```python
class Stock : 
    def __init__(self,name,data) : 
        self.name=name
        self.data=data
        
    def visualization(self):
        fig2 = make_subplots(specs=[[{"secondary_y": True}]])
        fig2.add_trace(go.Scatter(x=data['date'].index,y=data['closing'],name='Price'),secondary_y=True)
        fig2.add_trace(go.Bar(x=data['date'].index,y=data['volume'],name='Volume'),secondary_y=False)
        fig2.show()
    
    def max_value(self):
        return self.data['closing'].max()
    
    def min_value(self):
        return self.data['closing'].min()
    
    def max_volume(self):
        return self.data['volume'].max()
    
    def max_quantity(self):
        return self.data['quantity'].max()
    
    def momentum(self,N):
        l_moment=[]
        for i in N:
            momentum = data['closing'][731] - data['closing'][731-i]
            l_moment.append(momentum)   
        return l_moment   
    
    def simple_moving_average(self, window_size):
        return self.data['closing'].rolling(window=window_size).mean()
    
    def visualization2(self):
        fig2 = make_subplots(specs=[[{"secondary_y": True}]])
        fig2.add_trace(go.Scatter(x=data['date'].index,y=data['closing'],name='Price'),secondary_y=True)
        fig2.add_trace(go.Bar(x=data['date'].index,y=data['volume'],name='Volume'),secondary_y=False)
        fig2.add_trace(go.Line(x=data['date'].index,y=self.simple_moving_average(10),name='SMA'),secondary_y=True)
        fig2.add_trace(go.Line(x=data['date'].index,y=self.momentum(np.random.randint(low=1,high=730, size=700)),name='Momentum'),secondary_y=True)
        fig2.show()
        
```

```python
stock1 = Stock("bourse", data)
stock1.visualization2()
```
